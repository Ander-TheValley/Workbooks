# Dataframes

Los dataframes es el tipo de estructura de datos más importante en Pandas. Podemos pensar en ellos como una tabla o matriz, que estará formada por tanto por N filas y M columnas.

La función DataFrame nos permite crear DataFrames. Solo necesitamos introducir los datos.


```python
#Introducimos los datos de cada fila
mi_tabla =  pd.DataFrame( [[50, "María"], [31, "Juan"], [18, "Silvia"]])
mi_tabla
```

```python
#Podemos darle nombre a las columnas con el parametro "columns"
pd.DataFrame( [[50, "María"], [31, "Juan"], [18, "Silvia"]], columns=["Edad", "Nombre"])
```

```python
#Es más típico crearlo con un diccionario, ya que nos permite directamente indicar también el nombre de las columnas.
mi_tabla = pd.DataFrame( {"Edad":[50,31,18], "Nombre": ["María", "Juan", "Silvia"]}  )
mi_tabla
```

```python
mi_tabla.Edad
```

```python
mi_tabla.Nombre
```

```python
mi_tabla.index
```

```python
Id = [1,2,3]
mi_tabla.index = Id
mi_tabla
```

#Leyendo datos de una base de datos
Python, y concretamente Pandas nos permite conectarnos a una base de datos y ejecutar código SQL para recuperar datos y guardarlos en un DataFrame.

Para este ejemplo, utilizaremos la libreria sqlite que creará una pequeña base de datos a la que nos podremos

```python
####################  ESTE TROZO DE CÓDIGO NOS CREARÁ UNA PEQUEÑA BASE DE DATOS PARA UTILIZAR DE EJEMPLO. ####
####################   NO ES NECESARIO QUE LO APRENDÁIS, SOLO EJECUTARLO UNA VEZ                          ####
# Importamos SQLite
import sqlite3 
# conectamos a la BD
db_con = sqlite3.connect(r'database.db') 
# Leemos todos los csvs
c = pd.read_excel('https://raw.githubusercontent.com/tonilopezrosell/datasciencemasterthevalley/main/movies.xlsx')
o = pd.read_excel("https://raw.githubusercontent.com/tonilopezrosell/datasciencemasterthevalley/main/gender.xlsx")
r = pd.read_excel("https://raw.githubusercontent.com/tonilopezrosell/datasciencemasterthevalley/main/actors.xlsx")
p = pd.read_excel("https://raw.githubusercontent.com/tonilopezrosell/datasciencemasterthevalley/main/actor_movies.xlsx")
# Escribimos los txts como tablas en la base de datos.
c.to_sql('movies', db_con, if_exists='replace', index=False) 
o.to_sql('gender', db_con, if_exists='replace', index=False) 
r.to_sql('actors', db_con, if_exists='replace', index=False) 
p.to_sql('actor_movies', db_con, if_exists='replace', index=False) 

#Cerramos la conexion.
db_con.close()
```

```python
#Para conectarse a una base de datos necesitamos primero saber donde se encuentra, y normalmente usuario y contraseña. 
#Pero en esta ocasion nos basta con ejecutar la siguiente línea.
db_con = sqlite3.connect(r'database.db') 
```

```python
#Con la función read_sql de pandas podemos introducir código SQL y extraer datos de una base de datos.
#Solo necesitamos introducir dos argumentos: la consulta SQL, y la conexión a la base de datos.
actors = pd.read_sql("SELECT * FROM actors", db_con)
actors
```

```python
type(actors)
```

Recordad que podemos mezclar tablas con JOIN, para aprovechar y traernos ya los datos mezclados en un mismo dataframe.

```python
movies_and_genders = pd.read_sql("SELECT * FROM movies JOIN gender ON movies.genero = gender.genero", db_con)
movies_and_genders.head()
```

Ya sabemos conseguir datos de 3 maneras diferentes:


1.   Creando un dataframe nosotros mismos (con diccionarios o listas)
2.   Leyendo de ficheros CSV
3.   Conectandonos a una base de datos y ejecutando código SQL para traernos el resultado.

#Acceso condicional para filtrar 
Hasta ahora hemos aprendido los basics de acceder a diferentes filas y columnas de un dataset. En esta sección aprenderemos como filtrar filas basándonos en los valores que contienen.

```python
#Vereis que esta sentencia nos devuelve True o False para cada fila si cumple o no la condición.
movies.mov_year == 1999
```

```python
#Si pasamos el anterior trozo de código a loc, nos devolverá las filas que cumplen la condición.
movies.loc[movies.mov_year==1999]
```

```python
#Podemos combinar y crear condiciones tan complicadas como necesitemos con & (Y) y | (O)
movies.loc[ (movies.mov_year==1958) & (movies.mov_lang=="English")]
```

Existen varias funciones que nos facilitan la vida a la hora de seleccionar filas.



*   isin nos permite especificar más de un valor válido
*   notnull nos permite seleccionar las filas que no tengan valores nulos



```python
#para seleccionar todas las filas cuyo valor de mov_rel_country cumpla uno de estos dos valores
movies.loc[movies.mov_rel_country.isin(["JP", "USA"])]
```

```python
movies.loc[movies.mov_rel_country.isin(["JP", "USA"])==False] #podemos hacer la inversa añadiendo == False
```

```python
#todas las filas cuya region no sea null
movies.loc[movies.mov_rel_country.notnull()] #No sale la 6
```

# Asignando valores y creando nuevas columnas
Hasta ahora hemos aprendido como juguetear con los datos que hemos leído. En esta seccion aprenderemos como creamos nuevas columnas para ir aumentando y enriqueciendo nuestro DataFrame.



Podemos asignar un valor igual a todas las filas del DataFrame de una manera muy sencilla. En ocasiones queremos añadir una columna nueva que contenga un valor.

```python
movies["Formato"] = "Largometraje"
movies
```

En otras ocasiones queremos que ese valor solo se aplique cuando se cumpla una condicion. En ese caso combinaremos lo aprendido en la sección anterior.

```python
movies.loc[movies.mov_time > 200, "Formato"] = "Muy largometraje"
movies
```

Podemos cambiar todos los valores que cumplen una condición o varias combinando los operadores.

---



```python
movies.loc[(movies.mov_year > 1989) & (movies.mov_year < 2000), "mov_year"] = "90's"
movies
```

Podemos crear columnas como fórmulas matemáticas de otras columnas.


```python
movies["mov_time_in_secs"] = movies["mov_time"] * 60
movies.head()
```
